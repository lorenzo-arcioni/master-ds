In questo notebook ci occuperemo di un particolare tipo di dati: **serie temporali**. Iniziamo con una definizione di base su cui torneremo più avanti.

<h3><center><i>Una serie temporale è una sequenza di dati misurati nel tempo.</i></center></h3>

<br>

Fondamentalmente, ogni quantità che varia nel tempo è una serie temporale. Qualche esempio:

- temperatura minima giornaliera
- vendite totali di un prodotto
- livello di inquinamento atmosferico
- produzione e consumo di energia
- traffico web
- ...le infezioni della tua pandemia preferita.

Analisi vs previsione
---

Ripartiamo dalla definizione di serie temporali.

<h3><center><i>Una serie temporale è una sequenza di dati misurati nel tempo.</i></center></h3>

<br>


Le serie temporali forniscono dati con una relazione temporale.

Chiamiamo **analisi** il processo di comprensione e descrizione di una serie temporale di dati registrati in passato; tale analisi non è necessariamente focalizzata sulla previsione del futuro.
Quando utilizziamo esplicitamente una serie temporale per predire il suo futuro, stiamo **previsione**.

Continuiamo con l'analisi quantitativa e grafica. Nelle prossime lezioni, svilupperemo questi strumenti per fare previsioni.

Univariato vs multivariato
---

Le serie temporali **univariate** sono costituite da un'unica variabile osservata nel tempo, come il numero di vendite di un determinato prodotto o la temperatura minima giornaliera. Il set di dati che abbiamo caricato all'inizio di questa classe è un altro esempio di serie temporali univariate.

<img src='uni.png' width=700/>

<br>

In una serie temporale **multivariata**, quantità multiple variano nel tempo. Ad esempio, pensa a un sistema fisico in cui la velocità lungo i tre assi $x$, $y$ e $z$ viene registrata ad ogni passo temporale. Il set di dati **OPSD_Germany_all** è una serie temporale multivariata.


<img src='multi.png' width=700/>

<br>



Pari vs irregolare; 
---

Una serie temporale può avere una frequenza fissa (giornaliera, settimanale, ...) o variabile. Sono rispettivamente denominate serie temporali *pari* e *dispari*.

La frequenza delle serie temporali con cui abbiamo giocato finora è pari, giornaliera.

Vedremo come Panda gestisce la frequenza di una serie temporale. Iniziamo con *date_range*, che permette di creare un indice temporale con una data frequenza.

```python
pd.date_range('2020-03-09', '2020-05-04', freq='D')
```

```python
pd.date_range('2020-03-09', '2020-05-04', freq='W')
```

```python
pd.date_range('2020-03-09', '2020-05-04', freq='W-MON')
```

```python
pd.date_range('2020-03-09', '2020-05-04', freq='1H')
```

```python
df = pd.read_csv('OPSD_Germany_consumption.csv',
                 parse_dates=True,
                 index_col='Date')

df.head(10)
```

```python
df.asfreq('M').head(20)
```

Stagionalità
===

La tendenza di una variabile a ripetere periodicamente il proprio comportamento è chiamata **stagionalità**.

La stagionalità (annuale) è semplice con l'energia solare: ogni anno osserviamo lo stesso comportamento, a causa della disponibilità di luce solare.

La stagionalità non è necessariamente meteorologica: pensate ai saldi durante le vacanze di Natale.

Graphical analysis
===

Let's start our time series analysis using Python data visualization tools.

Matplotlib and Seaborn
---

<a href="https://matplotlib.org/">Matplotlib</a> is a highly customizable library, though a bit tricky at first. <a href="https://seaborn.pydata.org/.">Seaborn</a> standardises some graphical aspects and easily integrates with Pandas.

```python
import matplotlib.pyplot as plt

df = pd.read_csv('OPSD_Germany_all.csv',
                 parse_dates=True,
                 index_col='Date')

df.head()
```

```python
df_solar = df['Solar']

df_solar.plot()
```

```python
df_solar.plot(figsize=(15, 3), title='Solar energy production (GWh)')
```

```python
df_solar.plot(figsize=(15, 3), title='Solar energy production (GWh)', linewidth=.5)
```

```python
df_solar.plot(figsize=(15, 3),
              title='Solar energy production (GWh)',
              marker='.',
              markersize=3,
              linestyle='')
```

```python
df_solar['2014'].plot(figsize=(15, 3),
                      title='Solar energy production (GWh)',
                      marker='.',
                      markersize=3,
                      linestyle='')
```

```python
df_solar['2014-01'].plot(figsize=(15, 3),
                         title='Solar energy production (GWh)',
                         marker='.',
                         markersize=3,
                         linestyle='-')
```

Multiple plots
---

Let's try to *simultaneously* visualise solar and wind energy production.

```python
df_renewable = df[['Solar', 'Wind']] # sintassi per avere due colonne del dataset

df_renewable.plot(figsize=(15, 3),
                  title='Renawable energy production (GWh)',
                  marker='.',
                  markersize=3,
                  linestyle='')
```

```python
subplots = df_renewable.plot(figsize=(15, 6),
                             marker='.',
                             markersize=3,
                             linestyle='',
                             title=['Solar energy production', 'Wind energy production'],
                             subplots=True)

subplots
```

Pandas plot function is nothing but a shortcut to that of Matplotlib. For some specific graphical results (like the following) it is preferable to use the latter.

```python
plt.plot(df['Solar']['2013'],
         linestyle='',
         marker='.',
         markersize=3)
```

```python
plt.figure(figsize=(15, 5))

plt.plot(df['Solar']['2013'],
         linestyle='',
         marker='.',
         markersize=3)
```

```python
plt.figure(figsize=(15, 5))

plt.plot(df['Solar']['2013'],
         linestyle='',
         marker='.',
         markersize=3)

plt.plot(df['Wind']['2013'],
         linestyle='-',
         linewidth=0.5)
```

```python
plt.figure(figsize=(15, 5))

plt.plot(df['Solar']['2013'],
         linestyle='',
         marker='.',
         markersize=3,
         label='Solar')

plt.plot(df['Wind']['2013'],
         linestyle='-',
         linewidth=0.5,
         label='Wind')

plt.title('Renewable energy production (GWh)')
plt.legend()
```

**Hint**: think of each cell as an independent graphical environment.

It takes some time to become familiar with Matplotlib. However,

- it is a really versatile tool
- the Web is full with examples.

**Recap**: we have seen how to plot the content of a Dataframe with Pandas. We used Matplotlib to create sequential plots, wherein each subplot has specific graphical properties.
