$$H(X_{t}) := -\sum_{c=1}^{c}p_{ct}\log(p_{ct})$$

```python
# TO DO definire entropia e considerare il caso in cui esiste
# la forma indeterminata 0*-inf per capire 
#

def entropy(y):
    #modificata per risolvere il problema della forma indeterminata
    p = np.bincount(y.astype(int), minlength=2)/len(y)
    if np.sum(np.isclose(p, 0)) > 0: #ci torniamo sabato prossimo
        e = 0
    else:
        e = - np.sum(p * np.log2(p))# forma indeterminata con 0*-inf
    return e
```

```python
#TO DO calcolo del primo split (costruire f e soglia)

# Questa parte di codice serve per rappresentare un SOLO split
dec_list = []
queries = []

entropy_root = entropy(y_train) #entropia della radice
for j in range(X_train.shape[1]): # Itero sulle features (sugli indici delle features)
    s_list = np.unique(X_train[:,j]) # Prendo una sola volta tutte le occorrenze
    for s in s_list: # Per ogni tipo di occorrenza
        y_left = y_train[X_train[:,j] <= s]  # Calcolo come divide l'insieme 
        y_right =  y_train[X_train[:,j] > s] # del target
        
        n_l = len(y_left)
        n_r = len(y_right)
        n = len(y_train)
        
        if (n_l > 0  and n_r > 0): #per evitare di avere sottospazi con nessun campione #modificabile
            
            e_left = entropy(y_left)   # Calcolo l'entropia degli
            e_right = entropy(y_right) # insiemi generati dallo split secondo la feature j
            
            #decremo di impurità
            dec = entropy_root - (n_l / n * e_left +  n_r /n * e_right) # Calcolo la differenza di impurità 
            dec_list += [dec]
            queries += [{"feature":j, "soglia":s, "e_left":e_left, "e_right":e_right, "nl":n_l, "nr":n_r}]
```

```python
i = np.argmax(dec_list) # Ottengo la feature che più decrementa l'impurità
print(queries[i])
```
