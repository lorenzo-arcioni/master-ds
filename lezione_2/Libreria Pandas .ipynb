```python
dfr = dfr.drop(523, axis=0)
```

```python
#Creazione intervalli età per generazioni
import numpy as np
bins = [19,24, 39, 55, 74,85]
names = ['19-24','24-39', '39-55', '55-74','74-85']

d = dict(enumerate(names, 1))

dfr['age'] = np.vectorize(d.get)(np.digitize((dfr.Q5.astype(int)), bins))
```

```python
#Creazione macroaree regioni

macroarea = {'north-west':'nord','south':'sud','centre':'centro','north-east':'nord','south':'sud','islands':'sud'}

dfr['macroarea'] = dfr['region'].map(macroarea)
```

## Come creo dei subset del dataset?

La sintassi da utilizzare è la seguente:

    df[condition(df)]

```python
domande
```

```python
condizione = dfr.Q4 == 'Femmina' 
```

```python
condizione
```

```python
df_femmine = dfr[condizione]
df_femmine
```

```python
mask = (dfr.Q4 == 'Femmina') & (dfr.Q8 == 'Lombardia')
```

```python
df_donne_lombarde = dfr[mask]
```

```python
df_donne_lombarde.Q7.value_counts(normalize=True)*100
```

### Costrutto groupby

La funzione *groupby* permettere di organizzare in gruppi un DataFrame.

In questo caso:

- Raggruppa prima sulla base del sesso
- Estrae la colonna Q6 e ne calcola la percentuale di risposta rispetto ai diversi gruppi

```python
dfr.groupby('age')['Q6'].value_counts(normalize=True)*100
```
