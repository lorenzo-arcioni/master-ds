# Obiettivo: confrontare l'effetto di diversi scaler sui dati con valori anomali ¶

### In particolare confronta gli effetti dei tre scaler: MinMaxScaler, StandarScaler e RobustScaler sul dataset 'california_housing_dataset' presente nella libreria sklearn (https://scikit-learn.org/stable/datasets/real_world.html#california-housing-dataset)

#### Consiglio: graficare il pre e post scalatura può aiutarti ad osservare meglio come cambiano le distribuzioni ;)

```python
import numpy as np

import matplotlib as mpl
from matplotlib import pyplot as plt
from matplotlib import cm

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import minmax_scale
from sklearn.preprocessing import MaxAbsScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler

#DATI
from sklearn.datasets import fetch_california_housing

```

```python
df = fetch_california_housing(as_frame=True).frame
df = df.rename(columns=str.lower)
```

```python
df.head()
```

```python
#Plot della distanza 
plt.scatter(df["medinc"], df['aveoccup'])
plt.xlabel("Median income")
plt.ylabel("Average occupation")
```

### MinMax Scaling

```python
df_medinc_aveoccup_minmaxscaled = MinMaxScaler().fit_transform(df[['medinc', 'aveoccup']])
```

```python
#Plot della distanza 
plt.scatter(df_medinc_aveoccup_minmaxscaled[:, 0], df_medinc_aveoccup_minmaxscaled[:, 1])
plt.xlabel("Median income")
plt.ylabel("Average occupation")
```

### Standard Scaler

```python
df_medinc_aveoccup_standardscaled = StandardScaler().fit_transform(df[['medinc', 'aveoccup']])
```

```python
#Plot della distanza 
plt.scatter(df_medinc_aveoccup_standardscaled[:, 0], df_medinc_aveoccup_standardscaled[:, 1])
plt.xlabel("Median income")
plt.ylabel("Average occupation")
```

### Max Abs Scaler

```python
df_medinc_aveoccup_maxabsscaled = MaxAbsScaler().fit_transform(df[['medinc', 'aveoccup']])
```

```python
#Plot della distanza 
plt.scatter(df_medinc_aveoccup_maxabsscaled[:, 0], df_medinc_aveoccup_maxabsscaled[:, 1])
plt.xlabel("Median income")
plt.ylabel("Average occupation")
```

### Robust Scaling

```python
df_medinc_aveoccup_robustscaled = RobustScaler().fit_transform(df[['medinc', 'aveoccup']])
```

```python
#Plot della distanza 
plt.scatter(df_medinc_aveoccup_robustscaled[:, 0], df_medinc_aveoccup_robustscaled[:, 1])
plt.xlabel("Median income")
plt.ylabel("Average occupation")
```
